**For Loop**

Perulangan for digunakan untuk melakukan iterasi di dalam sequence seperti list, tuple, string, dan array atau objek yang dapat diiterasi lainnya.

**For Loop Syntax**

    for val in sequence:
        loop body

**Example: Python for Loop**

```python
#program untuk menghitung jumlah semua angka dalam sebuah list
L1 =[2, 4, 5, 7, 9]
jumlah = 0
#iterasi di dalam list
for i in L1:
  jumlah = jumlah + i
print(jumlah)
```

**The range() function**

```python
#program untuk melakukan iterasi di dalam list melalui indeks
minuman = ["teh", "boba", "kopi", "susu"]
p = len(minuman)

#iterasi di dalam list melalui indeks
for i in range(p):
  print(minuman[i])
```

**While Loop**
While loop pada Python digunakan untuk melakukan iterasi pada blok kode selama nilai dari test expression adalah true.

**syntax of while Loop in Python**

    while test_Expression:
        Body of while

  **while loop flowchart**

  psds2.png

  **example:python while loop**
  

```python
# program untuk menjumlahkan
# program asli hingga suku ke-n
# jum - 1+2+3+...+n

n = 10
jumlah = 0
i = 1

while i<= n:
  jumlah = jumlah + i
  i = i +1

print(jumlah)
```

**while loop with else**
sama dengan for loop kita juga dapat mengombinasikan while loop dengan blok else.
bagian else akan dieksekusi jika nilai pada . perhatikan contoh berikut:


```python
jum = 0
i = 1
while i <= n:
  jum = jum + i
  i = i + 1
else:
  print(jum) #dieksekusi siluar looping
```

Buatlah sebuah fungsi "Mahal Diskon dengan ketentuan:

* Input berupa data harga produk dengan tipe data numerik Tunggal

* Kategori input adalah "Murah" dan "Mahal", kemudian diskon ada 3 kategori "10%", "20%", dan "40%"

* Output Pertama sebuah teks/string yang menyatakan "Mahal" atau "Murah dan Keterangan "Kamu Dapat Diskon..


* Output Kedua adalah harga yang perlu dibayarkan

```python
# ketikan codingan anda
harga = float(input("masukkan harga: "))
if harga >= 100000:
    print("Mahal")
    if harga >= 100000 and harga < 200000:
      diskon = "10%"
      harga_diskon = harga - (harga * 0.1)
    elif harga >= 200000 and harga < 500000:
       diskon = "20%"
       harga_diskon = harga - (harga * 0.2)
    else:
      diskon = "40%"
      harga_diskon = harga - (harga * 0.4)
      print("Kamu Dapat Diskon " , diskon)
else:
    print("Murah")
    harga_diskon = harga
    print("")
print("harga yang perlu dibayarkan", harga_diskon)
```
